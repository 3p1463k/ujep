```python
from __future__ import absolute_import, division, print_function, unicode_literals
import pathlib
import pandas as pd
import numpy as np

"""Load data """
df = pd.read_excel('/home/evzen/Downloads/HSC.xlsx')
```

```python
df1 = df.iloc[:, [4,36]].drop(0).dropna()
"""Create list of dates"""
df1["Start date "] = df1["Start date "].astype(str)
dates_list = df1["Start date "].unique().tolist()
dates_list.remove("30.06.0219")
dates_list.append("2019-06-30 00:00:00")
#dates_list
vypocet_array = np.array([])

for x in dates_list:
    
    df2 = df1[df1["Start date "] == x]
    day_soucet = df2["Annotation Ratio RB/1500fr. [min]"].sum()
    day_pocet = df2["Annotation Ratio RB/1500fr. [min]"].shape[0]
    day_vypocet = day_soucet / day_pocet
    vypocet_array = np.append(vypocet_array, day_vypocet)

df4 = pd.DataFrame(data= {"Dates": dates_list, "Mean AR": vypocet_array})    
df4 = df4.sort_values("Dates")    
#df4.plot.line()
df4
```
