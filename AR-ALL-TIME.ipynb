```python
from __future__ import absolute_import, division, print_function, unicode_literals
import pathlib
import pandas as pd
import numpy as np
"""Load data """
df = pd.read_excel('/home/evzen/Downloads/HSC.xlsx')
"""Create dataset with name and Annotation ratio and fill NaN"""
df1 = df.iloc[:,[3,34]].drop(0, axis=0).fillna(0)
"""clean data and create list of names"""
df1 = df1[(df1 != 0).all(1)]
names_list =  df1.iloc[:,0]
nl = names_list.unique()
nl = np.delete(nl, -2)
nl = np.delete(nl, -10)
uniq_names = np.append(nl, "Vodák David")
my_list = uniq_names.tolist()
"""Calculate AR for each name in list"""
my_array = np.array([])
for x in my_list:
    filter_by_name = df1[df1["Annotator's name"]==x]
    soucet = filter_by_name.iloc[:, 1].sum()
    vypocet = soucet / len(filter_by_name)
    my_array = np.append(my_array, vypocet)
"""Create dataframe with names and calculations"""    
final_df = pd.DataFrame(data = {"Name": my_list,
                                 "AR":  my_array})
print("Prumer: " , final_df["AR"].mean())
print(final_df.describe())
final_df
```
