```python
from __future__ import absolute_import, division, print_function, unicode_literals
import pathlib
import pandas as pd
import numpy as np
"""Get input"""
input_date = input("Input date: ")
date = str(input_date)
"""Load data """
df = pd.read_excel('/home/evzen/Downloads/HSC.xlsx')

"""Filter columns and fill NaN and drop 0's"""
df1 = df.iloc[:,[3,34, 4]].drop(0, axis=0).fillna(0)
df1 = df1[(df1 != 0).all(1)]

"""Filter by date"""
df1["Start date "] = df1["Start date "].astype(str)
df2 = df1[df1["Start date "].str.contains(date)]

"""Create list of names for the day"""
names_list =  df2.iloc[:,0]
names_list = names_list.unique()
daily_names_list = names_list.tolist()
"""Calculate AR for each name in list and store as array"""
daily_list = np.array([])
final_df = pd.DataFrame()

for x in daily_names_list:
    name = x 
    """Create df with name"""
    df3 = df2[df2["Annotator's name"] == name]
    """store df to a list"""
    final_df = final_df.append(df3)
    """calculate AR and appen to list"""
    soucet= df3["Annotation Ratio NET/1500fr [min]"].sum()
    pocet = df3.shape[0]
    vypocet = soucet / pocet
    daily_list = np.append(daily_list, vypocet)
    
"""Create dataframe from calculated arrays and calculate AR"""
daily_ar_df = pd.DataFrame(data = {"Name": daily_names_list, "Daily mean": daily_list})
daily_ar_total_sum = daily_ar_df["Daily mean"].sum()
daily_pocet = daily_ar_df.shape[0]
daily_mean = daily_ar_total_sum / daily_pocet

"""Print to screen"""
print("Daily AR mean: " ,  daily_mean)
print(daily_ar_df.describe())
daily_ar_sorted = daily_ar_df.sort_values("Name")
daily_ar_sorted = daily_ar_sorted.rename(columns={"Annotator's name":"Name"})


"""Create df for each name with all values"""
daily_sorted = final_df.sort_values("Annotator's name")
daily_sorted = daily_sorted.rename(columns={"Annotator's name":"Name"})

"""Merge dataframes"""
merge_day = daily_ar_sorted.merge(daily_sorted)

"""Export to csv and excel"""
daily_ar_sorted.to_csv('/home/evzen/Downloads/Daily-AR-report-HSC.csv')
daily_ar_sorted.to_csv('/home/evzen/Downloads/Daily-AR-report-HSC.xlsx')
merge_day.to_csv('/home/evzen/Downloads/Daily-AR-detail-report-HSC.csv')
merge_day.to_csv('/home/evzen/Downloads/Daily-AR-detail-report-HSC.xlsx')
"""Display df"""
daily_ar_sorted
```
