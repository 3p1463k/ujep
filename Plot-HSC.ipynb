```python
from __future__ import absolute_import, division, print_function, unicode_literals
import pathlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline 
"""Load spreadsheet in to pandas dataframe"""
df = pd.read_excel('/home/evzen/Downloads/HSC.xlsx')
#df.isna().sum()
"""Clean the data, extract columns for analysis"""
df1 = df.iloc[:,-13:].drop(0,axis=0).drop(["Stav videa (info pro logistika a validtory)", 
                                          "Finish time - start time","as prce A v. pestvek (hod.)",
                                           "as prce V (hod.) efektivn", "Rozdl AQ vs. AO (kontrola pro anottora)",
                                          "Net time spent on video manipulation, clarifications, internal meetings (validtor)",
                                           "Annotation Ratio NET/1500fr [min] filter"], axis=1)
"""Fill NaN and drop outliers"""
df1 = df1.fillna(0).drop(1076)
"""plot dataframe"""
kernel_density = sns.pairplot(df1[["Annotation Ratio NET/1500fr [min]", "Annotation Ratio RB/1500fr. [min]",
                                   "Validation Ratio ", "as prce V (hod.) total",]], diag_kind="kde")
regression_line = sns.pairplot(df1[["Annotation Ratio NET/1500fr [min]", "Annotation Ratio RB/1500fr. [min]",
                                    "Validation Ratio ", "as prce V (hod.) total",]], kind="reg")

print(kernel_density)
print(regression_line)
```
